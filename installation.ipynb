## spacy.load("en_core_web_trf")
### There are two kind of model for now
### 1. en_core_web_trf
### 2. en_core_web_sm
### 1. en_core_web_trf has high accuracy and useful for production side (RoBERTa Transformer based).
### 2. en_core_web_sm has a bit low accuracy compared to en_core_web_trf but it's lightweight and mainly use for testing (CNN based)
### You can load pretrained language model using spacy.load().

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# We'll use tokenization for understand
sentence = nlp.tokenizer("We live in Paris and my name is Nouman Hafeez")
print(sentence)
```

```python
print("Tokens are:", len(sentence))
for i in sentence:
    print(i)
```
