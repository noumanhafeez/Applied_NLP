### There are two kind of model for now
### 1. en_core_web_trf
### 2. en_core_web_sm
### 1. en_core_web_trf has high accuracy and useful for production side (RoBERTa Transformer based).
### 2. en_core_web_sm has a bit low accuracy compared to en_core_web_trf but it's lightweight and mainly use for testing (CNN based)
### You can load pretrained language model using spacy.load().

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# We'll use tokenization for understand
sentence = nlp.tokenizer("We live in Paris and my name is Nouman Hafeez")
print(sentence)
```

```python
print("Tokens are:", len(sentence))
for i in sentence:
    print(i)
```

```python
# POS Tagging
doc = nlp("We live in Paris and my name is Nouman Hafeez.")
for token in doc:
    print(token.text, token.pos_, spacy.explain(token.dep_))
```

```python
# Visualize the dependency graph
from spacy import displacy

displacy.render(doc,  style='dep', jupyter=True, options={'distance': 120})
```

```python
# Chunking the related tokens into one
for chunks in nlp("Hello world. How are you?").noun_chunks:
    print(chunks.text)
```

```python
## Lemmatization
lemma = pd.DataFrame(data=[], columns=['original', 'lemma'])
i=0
for token in doc:
    lemma.loc[i, 'original'] = token.text
    lemma.loc[i, 'lemma'] = token.lemma_
    i = i+1

lemma
```

```python
eng = "George washington was an american actor and political leader"
print("Text Start Text End")
docs = nlp(eng)
for token in docs.ents:
    print(token.text, token.start_char, token.end_char, token.label_)
```

```python
displacy.render(docs, style='ent', jupyter=True, options={'distance': 120})
```
